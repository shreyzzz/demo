```python
import matplotlib.pyplot as plt
x=[368, 340, 665, 954, 331, 556, 376]
y=[1.7, 1.5, 2.8, 5, 1.3, 2, 1.3]
plt.scatter(x, y)
from scipy import stats
slope, intercept, r, p, std_error=stats.linregress(x, y)
def line(x):
  return slope*x+intercept
mymodel=list(map(line, x))
plt.plot(x, mymodel)
plt.show
```
